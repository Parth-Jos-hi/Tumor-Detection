```python
import os 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATASET_PATH = "C:/Users/pc/ML/Models/Tumor/Data"
YES_PATH = "C:/Users/pc/ML/Models/Tumor/Data/yes"
NO_PATH  = "C:/Users/pc/ML/Models/Tumor/Data/no"
```

```python
print(os.path.exists(DATASET_PATH))
print(os.path.exists(YES_PATH))
print(os.path.exists(NO_PATH))
```

```python
num_yes = len(os.listdir(YES_PATH))
num_no = len(os.listdir(NO_PATH))
print("Tumor images:", num_yes)
print("No-tumor images:", num_no)
```

```python
plt.bar(["Tumor", "No Tumor"], [num_yes, num_no])
plt.title("Class Distribution")
plt.show()
```

```python
def show_images(folder,title):
    images = os.listdir(folder)[:5]
    plt.figure(figsize= (10,4))
    for i ,img_name in enumerate (images):
        img = cv2.imread(os.path.join(folder,img_name),cv2.IMREAD_GRAYSCALE)
        plt.subplot(1,5,i+1)
        plt.imshow(img,cmap= "gray")
        plt.axis("off")
        plt.suptitle(title)
        plt.show()
show_images(YES_PATH,"Tumor Samples")
show_images(NO_PATH,"No Tumor Samples")
```

```python
sizes = []
for folder in[YES_PATH,NO_PATH]:
    for img_name in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,img_name),cv2.IMREAD_GRAYSCALE)
        sizes.append(img.shape)
sizes[:5]
```

```python
heights = [s[0] for s in sizes]
widths = [s[1] for s in sizes]
print("Min size:", min(heights), min(widths))
print("Max size:", max(heights), max(widths))
```

```python
img = cv2.imread(os.path.join(YES_PATH, os.listdir(YES_PATH)[0]), cv2.IMREAD_GRAYSCALE)

plt.hist(img.flatten(), bins=50)
plt.title("Pixel Intensity Distribution")
plt.show()
```

```python
img_resized = cv2.resize(img, (224, 224))
img_norm = img_resized / 255.0

plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Resized + Normalized")
plt.imshow(img_norm, cmap="gray")
plt.axis("off")

plt.show()
```
